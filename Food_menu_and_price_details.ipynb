```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://magicpin.in/New-Delhi/Paharganj/Restaurant/Eatfit/store/61a193/menu/"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
menu_section = soup.find('div', {'class': 'bricks-holder'})
```

```python
print(menu_section)
```

```python
items = menu_section.find_all('div', {'class': 'item-container'})
```

```python
print(items)
```

```python
for item in items:
    item_details = item.find('section', {'class': 'item-details'})
    name = item_details.find('p', {'class': 'item-name'}).text

    # Find the 'price-holder' class within 'item-details'
    price_holder = item_details.find('p', {'class': 'price-holder'})

    # Find the display price and actual price within 'price-holder'
    display_price = price_holder.find('span', {'class': 'display-price'}).text
    actual_price = price_holder.find('span', {'class': 'actual-price'}).text

    print(f"Item: {name}, Display Price: {display_price}, Actual Price: {actual_price}")
```
